```python
from selenium import webdriver
from bs4 import BeautifulSoup

def scrape_pinterest_tags(search_query):
    url = f"https://www.pinterest.com/search/pins/?q={search_query}"
    
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # Run Chrome in headless mode
    driver = webdriver.Chrome(options=options)
    
    driver.get(url)
    page_source = driver.page_source
    print(page_source)
    soup = BeautifulSoup(page_source, 'html.parser')
    text = soup.get_text()
    print(text)
    # Now you can find and extract the tag elements from the dynamically loaded content
    
    driver.quit()  # Close the browser instance

    # Rest of your scraping code

search_query = "fashion"
scrape_pinterest_tags(search_query)
```
